## 概要

transpileを用いると任意の回路を、指定したバックエンドの基本ゲートのみのゲートに変換できる。

まず、トフォリゲートを'ibmq_5_yorktown'の基本ゲートのみで表現する。


```python
from qiskit import IBMQ
from qiskit.circuit import *
from qiskit.compiler import transpile

# Loading your IBM Q account(s)
provider = IBMQ.load_account()
```

```python
device = provider.get_backend("ibmq_5_yorktown")

qc = QuantumCircuit(3)
qc.ccx(0, 1, 2)

qc_transpiled = transpile(qc, backend=device)
qc_transpiled.draw(output="mpl")
```

backendの指定ではなく、ゲートの指定でも実行できる

```python
qc = QuantumCircuit(3)
qc.ccx(0, 1, 2)

qc_transpiled = transpile(qc, basis_gates=["cx", "rz", "sx", "x"])
qc_transpiled.draw(output="mpl")
```

詳しくは[qiskit.compiler.transpile](https://qiskit.org/documentation/stubs/qiskit.compiler.transpile.html)を参照してください。
